```python
from PIL import Image
import os

def change_image_size_with_padding(source_dir, target_dir, target_height):
    # Ambil daftar nama file gambar dari direktori sumber
    img_list = os.listdir(source_dir)

    # Iterasi melalui setiap file gambar
    for img_name in img_list:
        img_path = os.path.join(source_dir, img_name)

        # Baca gambar menggunakan Pillow
        img = Image.open(img_path)

        # Tentukan dimensi baru
        target_width = img.width
        target_size = (target_width, target_height)

        # Buat gambar baru dengan padding untuk mencapai dimensi yang diinginkan
        new_img = Image.new("RGB", target_size, (0, 0, 0))  # Background putih
        new_img.paste(img, ((target_width - img.width) // 2, 0))

        # Tentukan path untuk menyimpan gambar yang sudah diubah ukurannya
        target_path = os.path.join(target_dir, img_name)

        # Simpan gambar yang sudah diubah ukurannya
        new_img.save(target_path)

# Tentukan direktori sumber gambar
source_directory = '/content/face-shape/diamond'

# Tentukan direktori target untuk menyimpan gambar yang sudah dihasilkan
target_directory = '/content/result'

# Tentukan tinggi target (640x480 -> tinggi 480)
target_height = 320

# Panggil fungsi untuk mengubah ukuran gambar dengan padding
change_image_size_with_padding(source_directory, target_directory, target_height)
```

```python
from PIL import Image
import os

def add_canvas_and_resize(source_dir, target_dir):
    img_list = os.listdir(source_dir)

    for img_name in img_list:
        image_path = os.path.join(source_dir, img_name)
        image = Image.open(image_path)

        # Determine the larger dimension of the original image
        max_dimension = max(image.size)

        # Create a blank canvas with a white background
        canvas_size = (max_dimension, max_dimension)
        canvas = Image.new('RGB', canvas_size, 'white')

        # Paste the original image onto the canvas
        paste_position = ((max_dimension - image.size[0]) // 2, (max_dimension - image.size[1]) // 2)
        canvas.paste(image, paste_position)

        # Resize the canvas with the pasted image to 320x320 pixels
        resized_image = canvas.resize((320, 320))

        # Save the resized image to the target directory
        target_path = os.path.join(target_dir, img_name)
        resized_image.save(target_path)

# Specify your source and target directories
source_directory = '/content/result'
target_directory = '/content/result'

# Call the function to add canvas and resize images
add_canvas_and_resize(source_directory, target_directory)
```

```python
from PIL import Image
import os

def resize_and_put_on_canvas(source_dir, target_dir, name):
    img_list = os.listdir(source_dir)

    for img_name in img_list:
        image_path = os.path.join(source_dir, img_name)
        image = Image.open(image_path)

        if image.mode == 'L':
          image.convert('RGB')

        # Resize the image to 320x320 pixels
        max_dimension = max(image.size)

        # Create a blank canvas with a white background
        canvas_size = (max_dimension, max_dimension)
        canvas = Image.new('RGB', canvas_size, 'white')

        # Paste the original image onto the canvas
        paste_position = ((max_dimension - image.size[0]) // 2, (max_dimension - image.size[1]) // 2)
        canvas.paste(image, paste_position)

        # Resize the canvas with the pasted image to 320x320 pixels
        resized_image = canvas.resize((320, 320))

        # Create a blank canvas with a white background
        canvas_size = (320, 320)
        canvas = Image.new('RGB', canvas_size, 'black')

        # Paste the resized image onto the canvas
        paste_position = ((canvas_size[0] - resized_image.size[0]) // 2, (canvas_size[1] - resized_image.size[1]) // 2)
        canvas.paste(resized_image, paste_position)

        new_img_name = img_name.replace(" ", "")

        # Save the final image to the target directory
        target_path = os.path.join(target_dir, new_img_name)
        canvas.save(target_path)

# Specify your source and target directories
#my_list = ["diamond", "heart", "oblong", "oval", "round", "square", "triangle"]
my_list = ["Heart", "Oblong", "Oval", "Round", "Square"]

for name in my_list:
  source_directory = '/content/datasets/' + name
  target_directory = '/content/result/' + name
  os.makedirs(target_directory, exist_ok=True)
  # Call the function to resize and put images on canvas
  resize_and_put_on_canvas(source_directory, target_directory, name)
```

```python
from PIL import Image
import os

def resize_and_put_on_canvas(source_dir, target_dir, name):
    img_list = os.listdir(source_dir)

    for img_name in img_list:
        try:
            image_path = os.path.join(source_dir, img_name)
            image = Image.open(image_path)

            if image.mode == 'L':
                image = image.convert('RGB')

            # Resize the image to 320x320 pixels
            max_dimension = max(image.size)

            # Create a blank canvas with a white background
            canvas_size = (max_dimension, max_dimension)
            canvas = Image.new('RGB', canvas_size, 'white')

            # Paste the original image onto the canvas
            paste_position = ((max_dimension - image.size[0]) // 2, (max_dimension - image.size[1]) // 2)
            canvas.paste(image, paste_position)

            # Resize the canvas with the pasted image to 320x320 pixels
            resized_image = canvas.resize((320, 320))

            # Create a blank canvas with a white background
            canvas_size = (320, 320)
            canvas = Image.new('RGB', canvas_size, 'black')

            # Paste the resized image onto the canvas
            paste_position = ((canvas_size[0] - resized_image.size[0]) // 2, (canvas_size[1] - resized_image.size[1]) // 2)
            canvas.paste(resized_image, paste_position)

            new_img_name = img_name.replace(" ", "")

            # Save the final image to the target directory
            target_path = os.path.join(target_dir, new_img_name)
            canvas.save(target_path)
        except Exception as e:
            print(f"Error processing image {img_name}: {e}")

# Specify your source and target directories
my_list = ["Heart", "Oblong", "Oval", "Round", "Square"]

for name in my_list:
    source_directory = '/content/datasets/' + name
    target_directory = '/content/result/' + name
    os.makedirs(target_directory, exist_ok=True)
    # Call the function to resize and put images on canvas
    resize_and_put_on_canvas(source_directory, target_directory, name)
```
